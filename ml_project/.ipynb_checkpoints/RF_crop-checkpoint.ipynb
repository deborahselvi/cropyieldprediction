```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
import csv


url = "apy.csv"
raw_data = open(url)
# cropdata = np.dtype([('State_Name', 'U40'), ('District_Name', 'U40'), ('Crop_Year', np.intc), ('Season', 'U40'), ('Crop','U40'), ('Area', np.double), ('Production', np.double)]) 

# reader = csv.reader(raw_data, delimiter=',', quoting=csv.QUOTE_NONE)
# x = list(reader)
# crop_prod = np.array(x)
# print(crop_prod.shape)
# crop_prod= np.genfromtxt(raw_data, dtype=cropdata, delimiter=",", names=True)
crop_prod= pd.read_csv(url)
# print(crop_prod.info())
# crop_prod = crop_prod.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
# print(crop_prod.dtype)
print(crop_prod)
print("State name values: {}".format(crop_prod.District_Name.value_counts()))
# crop_prod['Area'] = crop_prod['Area'].astype(float)
# crop_prod['Production'] = crop_prod['Production'].astype(float)
crop_prod['Area'] = pd.to_numeric(crop_prod['Area'], errors='coerce')
crop_prod['Production'] = pd.to_numeric(crop_prod['Production'], errors='coerce')
# class_le = LabelEncoder()
# crop_prod['State_Name'] = class_le.fit_transform(crop_prod['State_Name'].values)
# crop_prod['District_Name'] = class_le.fit_transform(crop_prod['District_Name'].values)
# crop_prod['Season'] = class_le.fit_transform(crop_prod['Season'].values)
# crop_prod['Crop'] = class_le.fit_transform(crop_prod['Crop'].values)

# crop_prod['State_Name'] = class_le.fit_transform(crop_prod['State_Name'])
# crop_prod['District_Name'] = class_le.fit_transform(crop_prod['District_Name'])
# crop_prod['Season'] = class_le.fit_transform(crop_prod['Season'])
# crop_prod['Crop'] = class_le.fit_transform(crop_prod['Crop'])

print("Original features:\n",list(crop_prod.columns),"\n")
crop_prod_dummies = pd.get_dummies(crop_prod)
print("Features after get_dummies:\n",list(crop_prod_dummies.columns))

features = crop_prod_dummies.loc[:,'Crop_Year':'Crop_other oilseeds']

X =features.values
y = crop_prod_dummies['Production'].values
print("X.shape: {} y.shape: {}".format(X.shape, y.shape))

# print(crop_prod.info())
# print(crop_prod)
# # print("max val -- {}".format(crop_prod.max()))
# print("max val -- {}".format(crop_prod['Production'].max()))
# X = crop_prod.iloc[:, 0:5].values
# y = crop_prod.iloc[:, 6].values
# print(crop_prod.shape)
# # X = crop_prod[:, 0:5]
# # y = crop_prod[:, 6]
X = np.nan_to_num(X)
y = np.nan_to_num(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

print("X_train --> {}".format(X_train))
print("y_train --> {}".format(y_train))

tree = DecisionTreeRegressor(random_state=0)
tree.fit(X_train, y_train)

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# for gamma in [1,10]:
#     svr = SVR(gamma=gamma).fit(X_train_scaled,y_train)
#     plt.plot(line, svr.predict(line), label ='SVR gamma={}'.format(gamma))
# plt.plot(X[:, 0], y, 'o', c='k')
# plt.ylabel("Regression output")
# plt.xlabel("Input feature")
# plt.legend(loc="best")

# forest = RandomForestRegressor(n_estimators=100,random_state=0, n_jobs=-1, max_leaf_nodes=16)
# forest.fit(X_train_scaled, y_train)

# print("Acuracy on training set: {:3f}".format(forest.score(X_train, y_train)))
# print("Accuracy on test set: {:3f}".format(forest.score(X_test, y_test)))

print("Acuracy on training set: {:3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:3f}".format(tree.score(X_test, y_test)))
```
